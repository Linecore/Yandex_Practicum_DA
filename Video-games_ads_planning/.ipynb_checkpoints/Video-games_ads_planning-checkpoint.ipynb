```python
def about_data(data, df_columns_list = [], limits = 'yes', limit_rows = 100000, limit_columns = 50, extension = False): 
    """
    V 1.1
    """
#-------------------------------------------------------------------------------------------------–ö–æ—Ä—Ä–µ–∫—Ü–∏—è –≤—Ö–æ–¥–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    ##------------------------------------------------------------------------------------------------–ü—Ä–æ–≤–µ—Ä–∫–∞ —Ç–∏–ø–∞ –ø–µ—Ä–µ–¥–∞—á–∏ –∫–æ–ª–æ–Ω–æ–∫
    if type(df_columns_list) != list:
        df_columns_list = [df_columns_list]
        print('‚ö†Ô∏è Warning. The type of inputs at the field "df_columns_list" should be "list".‚úÖ Input value is converted to list now.')
    ##------------------------------------------------------------------------------------------------–ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –≤–≤–æ–¥–∞ limits
    limits = limits.lower()
    if limits == 'y':
        limits = 'yes'
    elif limits == 'n':
        limits = 'no'
    elif limits == 'r':
        limits = 'rows'
    elif limits == 'c':
        limits = 'columns'
    elif limits == 'rows' or limits == 'columns' or limits == 'yes' or limits == 'no':
        pass
    else:
        limits = 'yes'
        print('‚ùå Error. Incorrect input in the field "limits".‚úÖ Setting set to default.')
    ##------------------------------------------------------------------------------------------------–ü—Ä–æ–≤–µ—Ä–∫–∞ —Ç–∏–ø–∞ –∏ —Ä–∞–∑–º–µ—Ä–æ–≤ –æ–≥—Ä–∞–Ω–∏—á–∏—Ç–µ–ª–µ–π —Å—Ç—Ä–æ–∫/—Å—Ç–æ–ª–±—Ü–æ–≤ 
    try:
        limit_rows = round(limit_rows)
    except:
        print('‚ùå Error. Incorrect input in the field "limit_rows" (must be type "integer").‚úÖ Setting set to default.')
        limit_rows = 100000 
    try:
        limit_columns = round(limit_columns)
    except: 
        print('‚ùå Error. Incorrect input in the field "limit_columns" (must be type "integer").‚úÖ Setting set to default.')
        limit_columns = 50
    ###-----------------------------------------------------------------------------------------------–ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ —É—Å—Ç–∞–Ω–æ–≤–∫–∏ limits –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ä–∞–∑–º–µ—Ä–æ–≤ –æ–≥—Ä–∞–Ω–∏—á–∏—Ç–µ–ª–µ–π
    if limit_rows == 0 and limit_columns != 0:
        limits = 'columns'
    elif limit_rows != 0 and limit_columns == 0:
        limits = 'rows'
    elif limit_rows == 0 and limit_columns == 0:
        limits = 'no'
        print('‚ö†Ô∏è Warning. Column and row limits are set: "0". ‚úÖ The limits setting set to "No".')
#-------------------------------------------------------------------------------------------------–û—Ü–µ–Ω–∫–∞ –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    ##---------------–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ–±–æ–±—â–µ–Ω–∏—è
    name_columns = []
    nan_info = []
    dupls = []
    nuniq = []
    total_row = True
##---------------------------------------------------------------------------------------------–û–±—Ä–∞–±–æ—Ç–∫–∞ –∏ –æ—Ü–µ–Ω–∫–∞ –≤ —Å–æ–æ—Ç–≤–µ—Å—Ç–≤–∏–∏ —Å —Ç–∏–ø–æ–º
    if str(type(data)) == "<class 'pandas.core.frame.DataFrame'>":
        print('‚û°Ô∏è Entered data type: DataFrame')
        
        if len(df_columns_list) > 0:
            checked_columns_list = []
            for i, name in enumerate(df_columns_list):
                if name in data.columns:
                    checked_columns_list.append(name)
                else:
                    print(f"‚ùå Error. The column '{name}' at position {i+1} is not part of entered DataFrame. ‚úÖ Column was ignored.")
            if len(checked_columns_list) == 0:
                checked_columns_list = data.columns
                print("‚ö†Ô∏è Warning. All entered columns was ignored. ‚úÖ All available columns will be used.")
            data = data[checked_columns_list]
            if len(df_columns_list) == 1:
                total_row = False
            else:
                total_row = True   
                
        rows_size, columns_size = data.shape 

        if (limits == 'yes' or limits == 'rows') and rows_size > limit_rows:

            data = data.loc[data.index[:limit_rows]]
            print(f'‚úÖ Row limit set: {limit_rows}')
            
        if (limits == 'yes' or limits == 'columns') and columns_size > limit_columns:
            data = data[data.columns[:limit_columns]]
            print(f'‚úÖ –°olumn limit set: {limit_columns}')

        for column_name in data.columns:
            name_columns.append(column_name)
            nan_info.append(data[column_name].isna().sum())
            dupls.append(data[data[column_name].isna() == False][column_name].duplicated().sum())
            nuniq.append(data[column_name].nunique())
            
        cells = (data.shape[0]*data.shape[1])
        
        if extension:
            try:
                about_data_visual_extention(data, df_columns_list = [], limits = limits, limit_rows = limit_rows, limit_columns = limit_columns)
            except:
                print("‚ùå Error. Visual extention is not available")
                
    elif str(type(data)) == "<class 'pandas.core.series.Series'>":

        print('‚û°Ô∏è Entered data type: Series')
        if len(df_columns_list) > 0: 
            print("‚ùå Error. Series don't have columns. ‚úÖ Field 'df_columns_list' was passed.")
            
        if (limits == 'yes' or limits == 'rows') and len(data) > limit_rows:
            data = data.loc[data.index[:limit_rows]]  
            print(f'‚úÖ Row limit set: {limit_rows}')
          
        nan_info = data.isna().sum()
        name_columns.append('Series')
        dupls.append(data[data.isna() == False].duplicated().sum())
        nuniq.append(data.nunique())
        cells = data.shape[0]
        total_row = False
        
        if extension:
            try:
                about_data_visual_extention(data, df_columns_list = [], limits = limits, limit_rows = limit_rows, limit_columns = limit_columns)
            except:
                print("‚ùå Error. Visual extention is not available")
                
    elif data == "author":
        return print('‚û°Ô∏è Hi there! I wrote this code to help people do data analysis. Together we can achieve more!\n‚úÖ Want to help or discuss? My github:','https://github.com/Linecore')
    
    else: #For Other Types of Data Storage
        return print('‚ùå Error. Incorrect type of data entered. Pandas Dataframe or Series only for now')
#----------------------------------------------------------------------------------------------------------–°–±–æ—Ä –æ–±–æ–±—â—ë–Ω–Ω–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –≤ —Ñ—Ä–µ–π–º –¥–ª—è –≤—ã–≤–æ–¥–∞      
    info_df = pd.DataFrame(data = nan_info, index = name_columns, columns = ['total_nan'])
    info_df['%_nan'] = round((info_df['total_nan'] / data.shape[0])*100,2)
    info_df['duplicated'] = dupls
    info_df['nunique'] = nuniq
    info_df['non_null'] = data.shape[0] - info_df['total_nan']
    info_df = info_df.sort_values(by='total_nan')
    info_df['type'] = data.dtypes
    if total_row:
        info_df.loc[f'TOTAL: {len(info_df)}'] = [info_df[info_df['total_nan'] != 0]['total_nan'].count(),  
                                                 round((info_df['total_nan'].sum()/cells)*100,2),
                                                 data.duplicated().sum(),
                                                 data.shape[0] - data.duplicated().sum(),
                                                 data.dropna().shape[0],
                                                 info_df['type'].nunique()]
    return info_df
```

–ï–¥–∏–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã–π –∫—Ä–æ—Å—Å–ø–ª–∞—Ç—Ñ–æ—Ä–º–µ–Ω–Ω—ã–π –≤—ã–ø—É—Å–∫ –∏–≥—Ä —Å—Ç–∞–ª –ø–æ–ø—É–ª—è—Ä–µ–Ω –æ—Ä–∏–Ω—Ç–µ—Ä–æ–≤–æ—á–Ω–æ –ø–æ—Å–ª–µ 2002. –•–æ—Ç—è —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∏ –∏—Å–∫–ª—é—á–µ–Ω–∏—è: Halo 2 –≤—ã–ø—É—â–µ–Ω–∞ –Ω–∞ xbox –≤ 2004, –Ω–∞ PC –≤ 2007, –∞ –ø–µ—Ä–µ–∏–∑–¥–∞–Ω–∏–µ –Ω–∞ xbox one - –≤ 2014 –≥–æ–¥—É. –¢–∞–∫ –∂–µ –±—ã–ª–∏ –ø–µ—Ä–µ–∏–∑–¥–∞–Ω–∏—è —Å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º –Ω–∞–∑–≤–∞–Ω–∏–µ–º –ø–æ —Ç–∏–ø—É —Ç–æ–≥–æ –∂–µ Need For Speed Most Wanted –∏–ª–∏ –ø–æ—Ä—Ç–∏—Ä–æ–≤–Ω–∏–µ/–ø–µ—Ä–µ–∏–∑–¥–∞–Ω–∏–µ —Ç–æ–π –∂–µ –∏–≥—Ä—ã –Ω–∞ –∫–æ—Å–Ω–æ–ª–∏ –Ω–æ–≤–æ–≥–æ –ø–æ–∫–∞–ª–µ–Ω–∏—è. –ü–æ—ç—Ç–æ–º—É —Å—É–¥–∏—Ç—å –æ —Ç–æ—á–Ω–æ–º –≥–æ–¥–µ —Ä–µ–ª–∏–∑–∞ –±–µ–∑ –¥–∞—Ç –≤—ã—Ö–æ–¥–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º - –Ω–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç—å—Å—è –≤–æ–∑–º–æ–∂–Ω—ã–º. –î–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º - –ø—Ä—è–º—ã—Ö –∫–æ–Ω–∫—É—Ä–µ–Ω—Ç–æ–≤, –≤ "–∏—Ö" —ç–ø–æ—Ö–µ –æ–±—ã—á–Ω–æ —ç—Ç–æ –æ–¥–∏–Ω –∏ —Ç–æ—Ç –∂–µ –≥–æ–¥ —Ä–µ–ª–∏–∑–∞. –î–ª—è –∏–≥—Ä –∂–∞–Ω—Ä–∞ Sports –∏ Fighting —á–∞—Å—Ç–æ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã –±—ã–ª–∏ –Ω–∞–∑–≤–∞–Ω–∏—è —Å –≥–æ–¥–æ–º –ø–æ —Ñ–æ—Ä–º—É–ª–µ "–≥–æ–¥ –≤—ã–ø—É—Å–∫–∞ + 1 = –≥–æ–¥ –Ω–∞–∑–≤–∞–Ω–∏—è".  –î–∞—Ç—ã –±—ã–≤–∞—é—Ç —Ä–∞–∑–Ω–æ–≥–æ –≤–∏–¥–∞: –ø–æ–ª–Ω–æ–µ —á–µ—Ç—ã—Ä—ë—Ö–∑–Ω–∞–∫–æ–≤–æ–µ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –≥–æ–¥–∞ (2004), –¥–≤—É—Ö–∑–Ω–∞–∫–æ–≤–æ–µ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –¥–≤—É—Ö —Ü–∏—Ñ—Ä –≥–æ–¥–∞ ('98, 06), —à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–µ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –≥–æ–¥–∞ (2K6 = 2006). –ò–∑–≤–ª–µ—á—ë–º –¥–∞—Ç—ã —Å–æ–≥–ª–∞—Å–Ω–æ —ç—Ç–æ–π —Ñ–æ—Ä–º—É–ª–µ.

```python
def date_extractor (row): #–ü—Ä–æ–∫–∞—á–∞–ª–∏ —Ä–µ–≥—É–ª—è—Ä–∫–∏
    if row[3] != 'Sports' and row[3] != 'Fighting':
        match = re.search(r'\s\d\d\d\d', row[0])
        if pd.isnull(match) == False:
            match = int(match[0])
            if 1985 < match < 2016:
                match = match - 1
                return match
            else:
                return np.NaN
    elif row[3] == 'Sports' or row[3] == 'Fighting':
        match = re.search(r'\s\d\d\d\d', row[0])
        match_sort = re.search(r'\s\d\d\b', row[0]) #06
        match_encrypt = re.search(r'\s\d\w\d',row[0]) #2K6
        match_encrypt2 = re.search(r'\s\d\w\d\d',row[0]) #2K13
        if pd.isnull(match) == False:
            match = int(match[0]) - 1
            return match
        if pd.isnull(match_sort) == False:
            match = int(match_sort[0])
            if 85 <= match <= 99:
                match = match + 1900 - 1
                return match
            if 0 <= match <= 16:
                match = match + 2000 - 1
                return match
        if pd.isnull(match_encrypt) == False:
            match = match_encrypt[0] 
            #—Ç–∞–∫ –∫–∞–∫ —Ç–∞–∫–∞—è –º–æ–¥–∞ –Ω–∞ 2–ö5 –ø—Ä–∏—à–ª–∞ –ø–æ—Å–ª–µ 2000, —Ç–æ –Ω—É–∂–Ω–æ —É–∑–Ω–∞—Ç—å –ø–æ—Å–ª–µ–¥–Ω—é—é —Ü–∏—Ñ—Ä—É.
            last = int(match[3])
            match = last + 2000 - 1
            return match
        if pd.isnull(match_encrypt2) == False:
            match = match_encrypt[0] 
            #—Ç–∞–∫ –∫–∞–∫ —Ç–∞–∫–∞—è –º–æ–¥–∞ –Ω–∞ 2–ö14 –ø—Ä–∏—à–ª–∞ –ø–æ—Å–ª–µ 2000, —Ç–æ –Ω—É–∂–Ω–æ —É–∑–Ω–∞—Ç—å –ø–æ—Å–ª–µ–¥–Ω–∏–µ —Ü–∏—Ñ—Ä—ã.
            last = int(match[2:])
            match = last + 2000 - 1
            return match
data_work = data.query('year_of_release.isna() == True')        
processing = data_work.apply(lambda row: date_extractor(row), axis = 1)
processing = processing.dropna()
print('–ó–∞–ø–æ–ª–Ω–µ–Ω–æ —Å—Ç—Ä–æ–∫:',len(processing))
index = processing.index
for i in index:
    data.loc[i,'year_of_release'] = processing[i]
```

–¢–∞–∫ –∫–∞–∫ –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç—Å—è –∏–≥—Ä–∞, —Ç–æ –æ—Ü–µ–Ω–∫–∏ ESRB –º–æ–∂–Ω–æ –∑–∞–ø–æ–ª–Ω–∏—Ç—å –∏—Å—Ö–æ–¥—è –∏–∑ —Ç–µ—Ö –∂–µ –∏–≥—Ä –Ω–∞ –¥—Ä—É–≥–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö. –ó–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Ä–µ–π—Ç–∏–Ω–≥–µ ESRB –Ω–∞ –ø–µ—Ä–≤—ã–π –ø–æ–ø–∞–≤—à–∏–π—Å—è. –ù–∞ —Ä–∞–∑–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –º–æ–≥—É—Ç –±—ã—Ç—å —Ä–∞–∑–Ω—ã–µ —Ä–µ–π—Ç–∏–Ω–≥–∏, –Ω–æ —Ä–∞–∑–ª–∏—á–∏—è –≤ –æ—Ü–µ–Ω–∫–∞—Ö –ø–æ—è–≤–ª—è—é—Ç—Å—è —Å–∫–æ—Ä–µ–µ –∏–∑ —Å—É–±—å–µ–∫—Ç–∏–≤–Ω—ã—Ö –º–Ω–µ–Ω–∏–π —ç–∫—Å–ø–µ—Ä—Ç–æ–≤ ESRB –ø–æ —Ä–∞–∑–Ω—ã–º –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.

```python
x = data[np.logical_and(data['rating'].isna()==True, data['year_of_release'] >= 1994)]['name'].unique()
before = data['rating'].isna().sum()

for i in x:
    idx = data[data['name'] == i].index
    z = data[data['name'] == i].dropna(subset = ['rating'])
    if z.empty != True:
        data.loc[idx,'rating'] = data.loc[idx,'rating'].fillna(z['rating'].unique()[0])
print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–æ –∑–∞–º–µ–Ω: {before - data['rating'].isna().sum()}")
```

–°–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –≤ –∫–æ—Ç–æ—Ä–æ–π –ø–æ—Å—á–∏—Ç–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ü–µ–Ω–æ–∫ –∏ –ø—Ä–æ–¥–∞–∂ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –ø–æ –≥–æ–¥–∞–º. –ê —Ç–∞–∫ –∂–µ –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ü–µ–Ω–æ–∫ –Ω–∞—Ö–æ–¥—è—â–∏—Ö—Å—è –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ "—Å–∫–æ–ª—å–∫–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å" –∏ "—Å–∫–æ–ª—å–∫–æ –µ—Å—Ç—å".

```python
work_data = data.pivot_table(index = 'year_of_release', 
                             values = ['critic_score','user_score','rating','na_sales', 'eu_sales','jp_sales', 'other_sales'], 
                             aggfunc = {'critic_score' : 'count', 'user_score' : 'count', 'rating' : 'count','na_sales':'sum',
                                        'eu_sales': 'sum', 'jp_sales': 'sum','other_sales':['sum','count']})

work_data.columns = (['critic_score_count','eu_sales_sum','jp_sales_sum','na_sales_sum',
                      'total_rows','other_sales_sum','rating_count','user_score_count'])

work_data.plot(y = ['critic_score_count','rating_count','user_score_count','total_rows'], 
               figsize = (10, 5), xlabel = '–ì–æ–¥', ylabel = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ, —à—Ç',
               title = '–ö–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ü–µ–Ω–æ–∫ —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –ø–æ –≥–æ–¥–∞–º –Ω–∞—Ö–æ–¥—è—â–µ–µ—Å—è –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ', 
              label = ['–û—Ü–µ–Ω–æ–∫ –∫—Ä–∏—Ç–∏–∫–æ–≤','–†–µ–π—Ç–∏–Ω–≥–æ–≤ ESRB','–û—Ü–µ–Ω–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π','–¶–µ–ª–µ–≤–æ–π –ø–æ–ª–Ω—ã–π –æ–±—ä—ë–º']);
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º –∏—Å—Ç–æ—Ä—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –æ–±—â–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏–≥—Ä –∏ –ø—Ä–æ–¥–∞–∂.

```python
platform_games = pd.DataFrame(index = work_data.columns, columns = ['year_born','year_dead','total_years','total_games'])
for i in work_data.columns:
    load = work_data.loc[:,i]
    load = load.dropna()
    total_games = load.sum()
    index = load.index
    platform_games.loc[i,:] = [index[0], index[-1], index[-1] - index[0] + 1, load.sum()]
data_work = data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
platform_games = platform_games.join(data_work['total_sales'], how = 'left')
platform_games.sort_values(by = 'total_sales',ascending=False).plot.bar(y = 'total_sales', 
                                                                        figsize = (10, 5), 
                                                                        xlabel = '–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞', 
                                                                        ylabel = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ, –º–ª–Ω. —à—Ç', 
                                                                        title = '–û–±—ä—ë–º –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º',
                                                                        legend = False)
sns.despine()
plt.show()
```

```python
index = platform_games.sort_values(by = 'total_sales',ascending=False).head(6).index
processing = data.query('platform in @index')
processing.pivot_table(index = 'year_of_release', columns = 'platform', values = 'total_sales', aggfunc = 'sum').fillna(0).plot(figsize = (10,5),
                                                                                                                                title = "–û–±—ä—ë–º –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º",
                                                                                                                                xlabel = '–ì–æ–¥', 
                                                                                                                                ylabel = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–ø–∏–π, –º–ª–Ω. —à—Ç')
plt.grid(axis = 'x')
plt.legend(title='–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
sns.despine()
plt.show()
```

### [–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –æ—Ç–∑—ã–≤–æ–≤ –∏ –ø—Ä–æ–¥–∞–∂](#plan)
–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –≤–Ω—É—Ç—Ä–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º –æ—Ç–∑—ã–≤—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –∫—Ä–∏—Ç–∏–∫–æ–≤. –ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ—è–Ω–∏—è –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –º–µ–∂–¥—É –æ—Ç–∑—ã–≤–∞–º–∏ –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏.

```python
def corr_show (paltform):
    processing = data.query('platform == @paltform')
    users = processing.dropna(subset = ['user_score'])
    critics = processing.dropna(subset = ['critic_score'])
    users = users[['user_score','total_sales']]
    critics = critics[['critic_score','total_sales']]
    print(f'–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞ {paltform}\nusers: {users.shape[0]}\ncritics: {critics.shape[0]}\n', users.corr(),'\n', critics.corr())
    fig, ax = plt.subplots(1,2, figsize = (10,5))
    ax[0].set_title(f'{paltform} User Score')
    ax[1].set_title(f'{paltform} Critic Score')
    ax[0].set_ylabel('Total Sales')
    ax[0].set_xlabel('User Score')
    ax[1].set_xlabel('Critic Score')
    ax[0].scatter(users['user_score'], users['total_sales'], alpha = 0.7)
    ax[1].scatter(critics['critic_score'], critics['total_sales'], alpha = 0.7)
    plt.show()
```

```python
for i in work_data['platform'].unique():
    corr_show(i)
```

```python
def top5 (df, sales):   
    df = df.reset_index()
    df.loc[5, df.columns[0]] = "Other"
    df.loc[5, sales] = df.loc[5:, sales].sum()
    return df.loc[:5]
```
