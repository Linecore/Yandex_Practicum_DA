Entertainment Software Rating Board (ESRB) — негосударственная организация, основное направление деятельности — принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде.

**Рейтиги присваеваемые ESRB:**
- **EC** («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.
- **E** («Everyone») — «Для всех»: Содержание вполне подходит для всех возрастов (ранее с 6 лет). Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. **Первоначально "K-A"** ("Kids to Adults").
- **E10** («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года.
- **T** («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.
- **M** («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику.
- **AO** («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое.
- **RP** («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.
- **RP 17+** («Rating Pending») — «Рейтинг ожидается - примерно 17+»: Продукт отправлен ESRB на оценку, но примерное содержание близко к рейтингу «M». Данная маркировка была введена с 2021 г. и распространяется на демо или презентационный вариант.

Единовременный кроссплатформенный выпуск игр стал популярен оринтеровочно после 2002. Хотя существуют и исключения: Halo 2 выпущена на xbox в 2004, на PC в 2007, а переиздание на xbox one - в 2014 году. Так же были переиздания с одинаковым названием по типу того же Need For Speed Most Wanted или портировние/переиздание той же игры на косноли нового покаления. Поэтому судить о точном годе релиза без дат выхода платформ - не представляеться возможным. Для платформ - прямых конкурентов, в "их" эпохе обычно это один и тот же год релиза. Для игр жанра Sports и Fighting часто характерны были названия с годом по формуле "год выпуска + 1 = год названия".  Даты бывают разного вида: полное четырёхзнаковое наименование года (2004), двухзнаковое наименование последних двух цифр года ('98, 06), шифрованное наименование года (2K6 = 2006). Извлечём даты согласно этой формуле.

```python
def date_extractor (row): #Прокачали регулярки
    if row[3] != 'Sports' and row[3] != 'Fighting':
        match = re.search(r'\s\d\d\d\d', row[0])
        if pd.isnull(match) == False:
            match = int(match[0])
            if 1985 < match < 2016:
                match = match - 1
                return match
            else:
                return np.NaN
    elif row[3] == 'Sports' or row[3] == 'Fighting':
        match = re.search(r'\s\d\d\d\d', row[0])
        match_sort = re.search(r'\s\d\d\b', row[0]) #06
        match_encrypt = re.search(r'\s\d\w\d',row[0]) #2K6
        match_encrypt2 = re.search(r'\s\d\w\d\d',row[0]) #2K13
        if pd.isnull(match) == False:
            match = int(match[0]) - 1
            return match
        if pd.isnull(match_sort) == False:
            match = int(match_sort[0])
            if 85 <= match <= 99:
                match = match + 1900 - 1
                return match
            if 0 <= match <= 16:
                match = match + 2000 - 1
                return match
        if pd.isnull(match_encrypt) == False:
            match = match_encrypt[0] 
            #так как такая мода на 2К5 пришла после 2000, то нужно узнать последнюю цифру.
            last = int(match[3])
            match = last + 2000 - 1
            return match
        if pd.isnull(match_encrypt2) == False:
            match = match_encrypt[0] 
            #так как такая мода на 2К14 пришла после 2000, то нужно узнать последние цифры.
            last = int(match[2:])
            match = last + 2000 - 1
            return match
data_work = data.query('year_of_release.isna() == True')        
processing = data_work.apply(lambda row: date_extractor(row), axis = 1)
processing = processing.dropna()
print('Заполнено строк:',len(processing))
index = processing.index
for i in index:
    data.loc[i,'year_of_release'] = processing[i]
```

Заполнили 24 даты релиза согласно правилам. Остальное невозможно заполнить точно исходя из имеющихся данных.

Так как оценивается игра, то оценки ESRB можно заполнить исходя из тех же игр на других платформах. Заполним пропуски в рейтинге ESRB на первый попавшийся. На разных платформах могут быть разные рейтинги, но различия в оценках появляются скорее из субьективных мнений экспертов ESRB по разным платформам.

```python
x = data[np.logical_and(data['rating'].isna()==True, data['year_of_release'] >= 1994)]['name'].unique()
before = data['rating'].isna().sum()

for i in x:
    idx = data[data['name'] == i].index
    z = data[data['name'] == i].dropna(subset = ['rating'])
    if z.empty != True:
        data.loc[idx,'rating'] = data.loc[idx,'rating'].fillna(z['rating'].unique()[0])
print(f"Произведено замен: {before - data['rating'].isna().sum()}")
```

### [Корреляция отзывов и продаж](#plan)
Посмотрим, как влияют на продажи внутри платформ отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.

```python
def corr_show (paltform):
    processing = data.query('platform == @paltform')
    users = processing.dropna(subset = ['user_score'])
    critics = processing.dropna(subset = ['critic_score'])
    users = users[['user_score','total_sales']]
    critics = critics[['critic_score','total_sales']]
    print(f'Платформа {paltform}\nusers: {users.shape[0]}\ncritics: {critics.shape[0]}\n', users.corr(),'\n', critics.corr())
    fig, ax = plt.subplots(1,2, figsize = (10,5))
    ax[0].set_title(f'{paltform} User Score')
    ax[1].set_title(f'{paltform} Critic Score')
    ax[0].set_ylabel('Total Sales')
    ax[0].set_xlabel('User Score')
    ax[1].set_xlabel('Critic Score')
    ax[0].scatter(users['user_score'], users['total_sales'], alpha = 0.7)
    ax[1].scatter(critics['critic_score'], critics['total_sales'], alpha = 0.7)
    plt.show()
```

```python
for i in work_data['platform'].unique():
    corr_show(i)
```

### [Пользовательские рейтинги PC и Xbox ONE](#plan)

Так как обе платформы - "принадлежат" Microsoft. То стоит проверить одинаковые ли пользовательские рейтинги у них.
Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC - равны.
Альтернативная гипотеза: Они разные. Это гипотезы можно проверить двусторонним тестом.

```python
processing = work_data.query('platform in ["PC","XOne"]').dropna(subset = ['user_score'])
```

```python
fig, ax = plt.subplots(1,2,figsize = (10,5))
fig.suptitle('Сравнение данных платформ')
category = processing['platform'].value_counts().index
values = list(processing['platform'].value_counts())
ax[0].bar(category, values, color='tab:blue', width=0.6)
ax[0].set_ylabel('Количество игр, шт')
ax[0].set_xlabel('Тип платформы')
ax[0].grid(axis='y')
ax[1].pie(values, labels=category, autopct='%1.1f%%',colors=['tab:blue', 'skyblue'])
plt.show()
print(processing['platform'].value_counts())
```

```python
xone = processing[processing['platform'] == 'XOne']
pc = processing[processing['platform'] == 'PC']
xone['user_score'].sort_values().hist(bins = 20, color = 'tab:blue')
pc['user_score'].sort_values().hist(bins = 20, alpha = 0.5, color = 'skyblue')
plt.show()
```

Соотношение нормальное. Установим уровень статистической значимости в 5 %.

```python
alpha = 0.05

results = st.ttest_ind(xone['user_score'], pc['user_score'], equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

P-Value выше уровня статистической значимости. Не получилось отвергнуть нулевую гипотезу о равенсте средних пользовательских рейтингов Xbox One и PC. Получение различий в наблюдениях будут скорее случайны, чем закономерны. С вероятностью в 11% различие будет получено случайно, что выше порога статистической значимости.

### [Пользовательские рейтинги жанров Action и Sport](#plan)

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») - одинаковые.
Альтернативная гипотеза: Они разные. Это гипотезы можно проверить двусторонним тестом.

```python
processing = work_data.query('genre in ["Action","Sports"]').dropna(subset = ['user_score'])
```

```python
fig, ax = plt.subplots(1,2,figsize = (10,5))
fig.suptitle('Сравнение данных платформ')
category = processing['genre'].value_counts().index
values = list(processing['genre'].value_counts())
ax[0].bar(category, values, color='tab:blue', width=0.6)
ax[0].set_ylabel('Количество пользователей')
ax[0].set_xlabel('Тип подписки')
ax[0].grid(axis='y')
ax[1].pie(values, labels=category, autopct='%1.1f%%',colors=['tab:blue', 'skyblue'])
plt.show()
print(processing['genre'].value_counts())
```

```python
action = processing[processing['genre'] == 'Action'].sample(160)['user_score'] #sample
sports = processing[processing['genre'] == 'Sports']
action.sort_values().hist(bins = 20, color = 'tab:blue')
sports['user_score'].sort_values().hist(bins = 20, alpha = 0.5, color = 'skyblue')
plt.show()
```

```python
alpha = 0.05

results = st.ttest_ind(action, sports['user_score'], equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

P-Value значительно меньше установленного уровня статистической значимости. С вероятностью в 8 % наблюдения будут различны, что ниже порога статистической значимости. То есть различия будут статистически значимы. Средние пользовательские рейтинги жанров Action и Sports - вероятно будут разные. Полученная разница будет скорее закономерна, чем случайна.
